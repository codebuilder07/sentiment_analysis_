```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
r=requests.get("https://www.yelp.com/biz/tesla-dedham?osq=Tesla+Dealership")
```

```python
r.status_code
```

```python
r.text
```

```python
soup=BeautifulSoup(r.text,'html.parser')
```

```python
divs= soup.findAll(class_='raw__09f24__T4Ezm')
```

```python
reviews=[]
for div in divs:
    reviews.append(div.text)
```

```python
len(reviews)
```

```python
reviews=reviews[7:18]
print(reviews)
```

```python
len(reviews)
```

```python
reviews1=[]
for i,div in enumerate(reviews):
    reviews1.append(div.replace('\xa0',''))
for i in reviews1:
    print(i)
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.array(reviews1),columns=['reviews'])
```

```python
df.head()
```

```python
len(df['reviews'])
```

```python
df['wordscount']=df['reviews'].apply(lambda x: len(x.split()))
```

```python
df.head()
```

```python
df['charscount']=df['reviews'].apply(lambda x: len((x)))
```

```python
df.head()
```

```python
def AverageNum(x):
    words=x.split()
    return (sum (len(word) for word in words)/len(words))
```

```python
df['Average']=df['reviews'].apply(lambda x: AverageNum(x))
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
```

```python
stop_words=stopwords.words('english')
```

```python
len(stop_words)
```

```python
df['stopword_count']=df['reviews'].apply(lambda x: len([word for word in x.split() if word in stop_words]))
```

```python
df.head()
```

```python
df['stopword_rate']=df['stopword_count']/df['wordscount']
```

```python
df.head()
```

```python
df.sort_values(by="stopword_rate")
```

```python
df.describe()
```

# DATA CLEANING


```python
df['lowercase']=df['reviews'].apply(lambda x: " ".join(word.lower() for word in x.split()))
```

```python
df.head()
```

```python
df['punctuation']=df['lowercase'].str.replace('[^\w\s]', '')
```

```python
df.head()
```

```python
df['nostopwords']=df['punctuation'].apply(lambda x: " ".join(word for word in x.split() if word not in stop_words))
```

```python
df['nostopwords'].head()
```

```python
pd.Series(" ".join(df['nostopwords']).split())#.value_counts()[:30]
```

```python
other_stopwords=["put","get","call","try","one","number","way","put","hold","getting","new"]
```

```python
df['nootherstopwords']=df['nostopwords'].apply(lambda x: " ".join(word for word in x.split() if word not in other_stopwords ))
```

```python
df['nootherstopwords'].head()
```

```python
pd.Series(" ".join(df['nootherstopwords']).split()).value_counts()[:20]
```
